```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
p_xp_exact = pd.read_csv('/arc/home/aydanmckay/p_xp_1_exact.csv')
p_xp_asec = pd.read_csv('/arc/home/aydanmckay/p_xp_1_asec.csv')
```

```python
ra_exact = abs(p_xp_exact['RA_p']-p_xp_exact['RA(CaHK)_c'])
dec_exact = abs(p_xp_exact['DEC_p']-p_xp_exact['Dec(CaHK)_c'])
ra_arcsec = abs(p_xp_asec['RA_p']-p_xp_asec['RA(CaHK)_c'])
dec_arcsec = abs(p_xp_asec['DEC_p']-p_xp_asec['Dec(CaHK)_c'])
```

```python
plt.semilogy()
plt.title('Difference in Angles when Performing the Exact Matching')
ra_e = plt.hist(ra_exact,histtype='step',label='Difference in RA')
dec_e = plt.hist(dec_exact,histtype='step',label='Difference in Dec')
plt.xlabel('Difference in Degrees in RA and Dec')
plt.ylabel('Population of Bin')
plt.legend()
plt.savefig('exactmatchontophots.png')
plt.show()
plt.semilogy()
plt.title('Difference in Angles when Performing the RA and Dec Matching')
ra_a = plt.hist(ra_arcsec,histtype='step',label='Difference in RA')
dec_a = plt.hist(dec_arcsec,histtype='step',label='Difference in Dec')
plt.xlabel('Difference in Degrees in RA and Dec')
plt.ylabel('Population of Bin')
plt.legend()
plt.savefig('asecmatchontophots.png')
plt.show()
```
